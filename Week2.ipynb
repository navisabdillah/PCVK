```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,225),thickness=15)
plt.imshow(black_img)
```

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='HELLO',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```

```python
#1 Tampilkan image dalam channel Red-Green
img_rg = img.copy()
img_rg[:,:,2] = 0
plt.imshow(img_rg)
```

```python
# Tampilkan image dalam channel Green-Blue
img_gb = img
img_gb[:,:,0] = 0
plt.imshow(img_gb)
```

```python
#2 Tampilkan  image baris ke 10-100, kolom 10-100
img_k = img[10:100,10:100]
plt.imshow(img_k)
```

```python
#3 Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
plt.imshow(img[5:30])
```

```python
#4 Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
imageWithRectangle = cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,10))
cv.rectangle(imageWithRectangle, pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
cv.rectangle(imageWithRectangle, pt1=(200,200),pt2=(300,300),color=(255,255,255),thickness=10)
cv.rectangle(imageWithRectangle, pt1=(400,100), pt2=(500,400), color=(0,0,255), thickness = 10)
cv.rectangle(imageWithRectangle, pt1=(200,150), pt2=(100,200), color=(255,0,0), thickness = 10)
cv.rectangle(imageWithRectangle, pt1=(500,150), pt2=(120,100), color=(0,0,255),thickness=10)
plt.imshow(imageWithRectangle)
```

```python
#5 Tampilkan image dengan posisi terbalik!
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

```python
#6. Buat rectangle dan circle pada bagian wajah dari image kucingku!

img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(450,150), pt2=(250,260), color=(255, 0, 30), thickness = 10)
cv.circle(img, center=(490, 200), radius=90, color=(180, 105, 255), thickness = 30)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```
